```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
import random
from datetime import datetime, timedelta
from tqdm.auto import tqdm
```

```python
def get_page_soup(url):
    # Récupération du contenu HTML de la page
    response = requests.get(url)
    html_content = response.content

    # Analyse du contenu HTML avec BeautifulSoup
    return BeautifulSoup(html_content, 'html.parser')
```

```python
soup = get_page_soup("http://cohen.frank.free.fr/photos/helicoptere/helicosamu/helico_du_samu.htm")

# Recherche de l'élément <map> dans le HTML
map_element = soup.find('map')

# Récupération de tous les liens (<area>) dans l'élément <map>
samus = map_element.find_all('area')

# Parcours de tous les liens et affichage de leur attribut "href"
for link in samus:
    print(link.get('href'))
```

```python
samu_data = []
for link in samus:
    samu_nb = link.get('href').split('/')[-1][:-4]
    copter_soup = get_page_soup("http://cohen.frank.free.fr/photos/helicoptere/helicosamu/" + link.get('href'))

    # extraire les données de la table dans un dictionnaire
    # Find the table which has "Immatriculation" in its first cell
    table = None
    for t in copter_soup.find_all("table"):
        if len(t.find_all("tr")) > 0 and "Immatriculation" in t.find_all("tr")[0].find_all("td")[0].text.strip():
            table = t
            break

    if table:
        # Extract the data from the table
        data = {'samu': samu_nb}
        for row in table.find_all("tr"):
            cols = row.find_all("td")
            key = cols[0].text.strip().replace(" :", "").replace("\n", "").replace("\xa0", "").replace("\t", "").replace("        ", " ").replace("       ", " ").replace("      ", "")
            value = cols[1].text.strip()
            data[key] = value
        samu_data.append(data)
    else:
        print("No table found with 'Immatriculation' in its first cell.")
samu_data
```

```python
samu_df = pd.DataFrame.from_records(samu_data)
samu_df
```

# RadarBox scrapper

- https://www.radarbox.com/data/registration/F-HLCG
- https://www.radarbox.com/data/mode-s/39AC46
- https://www.radarbox.com/data/flights/SAMU61

Les 3 donnent en fait la même vue donc c'est plus simple d'attaquer directement avec le nom du SAMU !

```python
for _, copter in samu_df.head(2).iterrows():
    soup = get_page_soup("https://www.radarbox.com/data/flights/" + copter["samu"])
    status = soup.find("div", {'class': 'FlightStatus'})
    departure = soup.find("div", {"id": "departure"})
    arrival = soup.find("div", {"id": "arrival"})
    lat = soup.find("div", {"id": "position-info"}).find_all('div')[2]
    print({'status': status, 'departure': departure, 'arrival': arrival, 'lat': lat})
```

```python
# Noticed an API is called to fill lat / lon
keys_to_keep = ['alt', 'la', 'lo', 'lastlalot', 'status', 'departure', 'arrival', 'acr', 'acd', 'accn', 'ms', 'fid']

live_data = []
for _, copter in tqdm(samu_df.iterrows()):
    print(copter["samu"], end=" ")
    response = requests.get("https://www.radarbox.com/data/flights/info?type=flights&query=SAMU" + copter["samu"])
    if response.content != b'':
        data = response.json()
        subset_dict = {key: data[key] for key in keys_to_keep}
        subset_dict['samu'] = copter["samu"]
        live_data.append(subset_dict)
        print('-> done.')
    else:
        print('-> skipped.')
    time.sleep(random.uniform(0.3, 2.0))
live_df = pd.DataFrame.from_records(live_data)
```

```python
# Collect current flight info
def get_flight_data(samu, flight):
    # Livedata is forbidden
    # response = requests.get("https://data.rb24.com/live?flightid=1940723279")

    # Through log page (ex: https://www.radarbox.com/data/flights/SAMU51/1940723279/log)
    soup = get_page_soup(f"https://www.radarbox.com/data/flights/SAMU{samu}/{flight}/log")
    table = soup.find('table')

    # Get the column headers (th elements)
    headers = [th.text for th in table.find_all('th')]

    # Get the table rows (tr elements)
    rows = table.find_all('tr')[1:]  # skip the first row since it contains the headers

    # Create a list of dictionaries where each row in the table is represented as a dictionary
    flight_data = [dict(zip(headers, [td.text for td in row.find_all('td')])) for row in rows]
    return flight_data
```

```python
live_df['lastlalot'] = pd.to_datetime(live_df['lastlalot'], unit='ms')

def time_ago(dt):
    now = datetime.now()
    diff = now - dt
    if diff < timedelta(minutes=1):
        return 'just now'
    elif diff < timedelta(hours=1):
        return f'{diff.seconds // 60} minutes ago'
    elif diff < timedelta(days=1):
        return f'{diff.seconds // 3600} hours ago'
    else:
        return f'{diff.days} days ago'
live_df['time_ago'] = live_df['lastlalot'].apply(time_ago)

live_df['last'] = live_df.apply(lambda copter: get_flight_data(copter['samu'], copter['fid'])[-1])

live_df
```
